```python
import re
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('spam.csv')
df
```

```python
df.isnull().sum()
```

```python
def clean(text):
    text= text.lower()
    text= re.sub(r'[^\w\s]','', text)
    tokens = word_tokenize(text)
    tokens= [WordNetLemmatizer(). lemmatize(word) for word in tokens]
    tokens= [word for word in tokens if word not in stopwords.words('english')]
    return "".join(tokens)
```

```python
df['email']= df['email'].apply(clean)
```

```python
df
```

```python
x_train, x_test, y_train, y_test=train_test_split(df['email'],df['label'], test_size=0.2, random_state=42)
```

```python
pipeline=Pipeline([('tfidf', TfidfVectorizer()), ('Classifier',MultinomialNB())]) 
```

```python
pipeline.fit(x_train, y_train)
```

```python
user_email= input("Enter you email:")
cleaned_email= clean(user_email)
predict = pipeline.predict([cleaned_email])
print("Result:", predict)
```
